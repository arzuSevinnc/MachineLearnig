```python
import sys 
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
%matplotlib inline
```

```python
original = plt.imread('image.jpg')
plt.imshow(original)
```

```python
imageNew = Image.fromarray(np.uint8(original))
imageNew = ImageOps.equalize(imageNew)
plt.imshow(imageNew)
```

```python
from skimage.color import rgb2gray
```

```python
grayscale = rgb2gray(original)

fig, axes = plt.subplots(1, 2, figsize=(8, 4))
ax = axes.ravel()

ax[0].imshow(original)
ax[0].set_title("Original")
ax[1].imshow(grayscale, cmap=plt.cm.gray)
ax[1].set_title("Grayscale")

fig.tight_layout()
plt.show()
```

```python
img = np.array(Image.open('image.jpg'))

img_R = img.copy()
img_R[:, :, (1, 2)] = 0
img_G = img.copy()
img_G[:, :, (0, 2)] = 0
img_B = img.copy()
img_B[:, :, (0, 1)] = 0

imgRed = Image.fromarray(img_R)
imgRed.save('red.jpg')
imgGreen = Image.fromarray(img_G)
imgGreen.save('green.jpg')
imgBlue = Image.fromarray(img_B)
imgBlue.save('blue.jpg')

```
