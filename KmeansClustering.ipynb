```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
originalImg = plt.imread("image2.png")
plt.imshow(originalImg)
```

```python
image = cv2.imread("image2.png")
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
pixel_values = image.reshape((-1, 3))
pixel_values = np.float32(pixel_values)

print(pixel_values.shape)
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
```

```python
k = 3
_, labels, (centers) = cv2.kmeans(pixel_values, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
```

```python
centers = np.uint8(centers)

labels = labels.flatten()
```

```python
segmented_image = centers[labels.flatten()]
```

```python
segmented_image = segmented_image.reshape(image.shape)

plt.imshow(segmented_image)
plt.show()
```

```python
clusterImg = np.copy(image)
clusterImg = clusterImg.reshape((-1, 3))
clusterImg_2 = np.copy(image)
clusterImg_2 = clusterImg_2.reshape((-1, 3))
clusterImg_3 = np.copy(image)
clusterImg_3 = clusterImg_3.reshape((-1, 3))

cluster = 0
clusterImg[labels == cluster] = [0, 0, 0]

clusterImg = clusterImg.reshape(image.shape)

cluster_2 = 1
clusterImg_2[labels == cluster_2] = [0, 0, 0]

clusterImg_2 = clusterImg_2.reshape(image.shape)

cluster_3 = 2
clusterImg_3[labels == cluster_3] = [0, 0, 0]

clusterImg_3 = clusterImg_3.reshape(image.shape)

plt.imshow(clusterImg)
plt.show()
plt.imshow(clusterImg_2)
plt.show()
plt.imshow(clusterImg_3)
plt.show()
```
